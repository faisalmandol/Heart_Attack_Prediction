## Finding outliers

```python
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outlier

# Call the function *outside* the function block
chol_outliers = find_outliers_iqr(data, 'Cholesterol')
print("Cholesterol Outliers:", len(chol_outliers))
```

```python
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outlier

# List of numeric columns to check for outliers
columns_to_check = ['Cholesterol', 'BMI', 'BloodPressure', 'HeartRate', 'Age', 'MaxHeartRate', 'ST_Depression']

for col in columns_to_check:
    outliers = find_outliers_iqr(data, col)
    print(f"{col} -> Outliers: {len(outliers)}")

```

## **Processs Data**

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in data.select_dtypes(include = 'object'):
    data[col] = le.fit_transform(data[col])
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

x = data.drop('Outcome', axis = 1)
y = data['Outcome']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

## Model Train

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state=42, stratify = y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model = LogisticRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

train_score = accuracy_score(y_train, y_train_pred)
test_score = accuracy_score(y_test, y_test_pred)

print("Training Accuracy: {:.2f}".format(train_score))
print("Testing Accuracy: {:.2f}".format(test_score))
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
print("Accuracy: {:.2f}".format(accuracy_score(y_test, y_pred)))

# Confusion Matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Precision, Recall, F1-score
print("Classification Report:\n", classification_report(y_test, y_pred))
```
